```python
import requests as req
from bs4 import BeautifulSoup as bs
from urllib import parse
```

```python
# 取得新聞列表
url = "https://www.iiiedu.org.tw/category/news/" 

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url)

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml")
```

```python
# 取得 a 的文字
for a in soup.select("article h2.entry-title a"):
    print(a.get_text())
```

```python
# 建立 list 來放置新聞連結
list_news = []

# 取得 a 的連結
for a in soup.select("article h2.entry-title a"):
    list_news.append(a["href"])  # 塞入 list
    '''
    補充
    parse.quote(" ") => "%20"
    parse.unquote("%20") => " "
    '''
    
# 走訪每一個 a link，並印出網頁內文
for index, link in enumerate(list_news):
    res = req.get(link)
    soup_ = bs(res.text, "lxml")
    print(soup_.select_one("div.entry-content").get_text())
    print("=" * 75)
```

```python
 # PTT NBA 板
url = "https://www.ptt.cc/bbs/NBA/index.html"

res = req.get(url)

soup = bs(res.text, "lxml")
```

```python
for a in soup.select("div.r-ent div.title a"):
    print(a.get_text())
```

```python
list_article = []

for a in soup.select("div.r-ent div.title a"):
    list_article.append(a["href"])

for index, link in enumerate(list_article):
    res = req.get("https://www.ptt.cc" + link)
    soup_ = bs(res.text, "lxml")
    content = soup_.select_one("div#main-content").get_text()
    pre_content = content.split("--")[0]
    print(pre_content)
```
