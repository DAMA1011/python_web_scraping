```python
'''
參考網頁
[1] Python 使用 Beautiful Soup 抓取與解析網頁資料，開發網路爬蟲教學
https://blog.gtwang.org/programming/python-beautiful-soup-module-scrape-web-pages-tutorial/2/
'''

import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint

 # PTT NBA 板
url = "https://www.ptt.cc/bbs/NBA/index.html"

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url)

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml")
print(soup)
```

```python
# 第一個 <a></a>
print(soup.find("a"))
print("\n")

# 全部 <a></a>，此時回傳 list
pprint(soup.find_all('a'))
print("\n")

# 指定 list 某個元素的 html
print(soup.find_all('a')[2])
```

```python
# 取得 id 為 logo 的元素
logo = soup.find(id = "logo")
print(logo)
print("\n")

# 取得所有 div，類別名稱為 r-ent，回傳為 list
posts = soup.find_all("div", class_ = "r-ent")
print(posts)
```

```python
'''
以下透過 CSS selector 取得元素，
回傳格式為 list
'''

# 輸出 title
print(soup.select_one("title"))
print("\n")

# 輸出 a
print(soup.select("a"))
print("\n")

# 透過 class 名稱取得元素
print(soup.select("a.board"))  # a[.class名][.class名]
print("\n")

# 透過 id 名稱取得元素
print(soup.select_one("#logo"))
print("\n")

# 透過 attribute 取得元素
# <a herf="#" data-id="21">hello</a>
print(soup.select("a[class='board']"))  # a[class="class名 class名"]
```

```python
# 取得單一節點的文字內容 (select_one 會回傳單一 bs element 物件，select 會回傳 list)
print(soup.select_one("title").get_text())
print(soup.select_one("title").text)
print(soup.select("a")[0].get_text())
print(soup.select("a")[0].text)
```

```python
# 透過迭代取得所有 a 的文字內容
for a in soup.select("a"):
    print(a.get_text())
```

```python
# 透過迭代取得所有 a 的屬性 href
for a in soup.select("a"):
    if a.has_attr("href"):
        print("(https://www.ptt.cc)" + a["href"])  # a.get("herf")
    else:
        print("=" * 50)
        print(f"連結[{a.get_text()}] 沒有 href 屬性")
        print("=" * 50)
```
