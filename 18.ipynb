```python
'''
匯入工具
'''
import pyautogui
from time import sleep, time
import keyboard, mouse
from IPython.display import clear_output
```

```python
'''
請在 Terminal 當中執行
取得滑鼠座標和游標上的色碼，協助我們定位
補充: 必須在 Terminal 當中執行，請見 18.py
'''
# pyautogui.displayMousePosition()
```

```python
'''
取得目前滑鼠座標
'''
while True:
    # 自動清除此格的文字輸出
    clear_output(wait=True)
    
    # 取得滑鼠游標的座標，並加以輸出
    x, y = pyautogui.position()
    print(f'x={x}, y={y}')
    
    # 睡一下，以免吃光記憶體
    sleep(0.1)
```

```python
'''
記錄滑鼠軌跡
'''
# 紀錄軌跡用的變數
listMove = []

# 紀錄軌跡
def record():
    # 監聽事件
    while True:
        # 按下 Ctrl 錄製軌跡，放開就不錄製
        if keyboard.is_pressed('ctrl'):
            x ,y = pyautogui.position()
            listMove.append(f'{x}, {y}')
            sleep(0.1)
        elif keyboard.is_pressed('alt'):
            # 按下 alt 結束程式
            break
            
    # 寫入檔案
    with open('move.txt', 'w', encoding='utf-8') as file:
        file.write('\n'.join(listMove))

'''主程式'''
if __name__ == '__main__':
    record()
```

```python
'''
讀取滑鼠軌跡，並模擬滑鼠遊標移動
'''
# 模擬移動
def move():
    # 讀取軌跡檔
    with open('move.txt', 'r', encoding='utf-8') as file:
        # 逐行讀取
        for line in file:
            # 切割文字，找出 x, y
            x, y = line.strip().split(',')
            
            # 移動(設定移動時間)
#             pyautogui.moveTo(int(x), int(y), duration=0.1)
            pyautogui.rightClick(int(x), int(y), duration=0.1)
            

            # 提前結束
            if keyboard.is_pressed('alt'):
                break
            
'''主程式'''
if __name__ == '__main__':
    move()
```

```python
'''
每秒判斷桌面上的圖片(座標)是否存在
補充: 請先開小算盤(計算機)
'''
while True:
    # 等待
    sleep(0.1)
    
    # 自動清除此格的文字輸出
    clear_output(wait=True)
    
    # 取得 Box 物件
    btn_2_location = pyautogui.locateOnScreen('./images/2.png', confidence=0.9)
    
    # 輸出
    print(btn_2_location)
```

```python
'''
打擊遊戲

範例網址: https://taiko.bui.pm/

請搭配 18.py，取得參考用的座標、色碼
注意: 每台電腦的顏色配置可能不同，要自行找到合適的色碼

    Red, Green, Blue
橘色: (243, 71, 40)
藍色: (101, 189, 187)
連打: (243, 181, 0)
248, 119, 0

打擊區座標
(x=519, y=455)
'''
# 設定每一個動作，都暫停若干秒
pyautogui.PAUSE = 0.03

# 打擊點
listPoint = [
    [536, 421],
    [536, 436],
    [537, 459]
]

# 點擊
def click():
    for list_ in listPoint:
        # 每次點擊都會重新取得指定打擊點的 RGB 值
        coord = pyautogui.pixel(list_[0], list_[1])
        
        # coord[0] 代表 Red，coord[1] 代表 Green，coord[2] 代表 Blue
        if coord[0] == 243 and coord[1] == 71:
            pyautogui.press('j')
        elif coord[0] == 101 and coord[1] == 189:
            pyautogui.press('k')
        elif coord[0] == 243 and coord[1] == 181:
            pyautogui.press('k')
        elif coord[0] == 248 and coord[1] == 119:
            pyautogui.press('d')
            
            
'''主程式'''
if __name__ == '__main__':
    while True:
        # 當我們按下 ctrl 鍵時，自動點擊，按下 alt 則結束迴圈
        if keyboard.is_pressed('ctrl'):
            click()
        elif keyboard.is_pressed('alt'):
            break
```
